# PulseML — Exploratory Data Analysis
Column meanings, missing-value rates, and target distribution for the three raw tables.

```python
import pandas as pd

app = pd.read_csv('../data/raw/application_train.csv')
bureau = pd.read_csv('../data/raw/bureau.csv')
prev = pd.read_csv('../data/raw/previous_application.csv')
print(app.shape, bureau.shape, prev.shape)
```

## Target distribution
Class imbalance matters for both metric choice (AUC-ROC + AUC-PR, not accuracy) and model config (`scale_pos_weight`).

```python
app['TARGET'].value_counts(normalize=True)
```

## Missing-value rates per table

```python
for name, df in [('application_train', app), ('bureau', bureau), ('previous_application', prev)]:
    print(name)
    print((df.isna().mean() * 100).round(2))
    print()
```

## Column meanings
- `AMT_INCOME_TOTAL` / `AMT_CREDIT` / `AMT_ANNUITY`: declared income, requested credit, and periodic repayment amount on the application.
- `DAYS_EMPLOYED`: negative = days employed before the application date.
- `bureau.CREDIT_ACTIVE`: whether a prior credit-bureau record is still open (`Active`) or `Closed`.
- `previous_application.NAME_CONTRACT_STATUS`: outcome of a prior loan application with this lender (`Approved`/`Refused`/`Cancelled`).

See `src/features/feature_definitions.py` for the full engineered-feature data dictionary.
